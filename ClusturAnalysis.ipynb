```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv files
data = pd.read_csv(r"C:\Users\elise\OneDrive\Documenten\Data_science\Mall Customer Segmentation Data\Mall_Customers.csv")
```

```python
data.head()
```

shuffle data

```python
mail_data = data.sample(frac=1, replace=False)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = "Annual Income (k$)", 
                y = "Spending Score (1-100)", 
                data = mail_data,
               hue = "Age")
plt.show()
```

# Analyse data

### group by age

```python
age = mail_data.groupby("Age").mean().reset_index()
age.head()
```

```python
plt.figure(figsize=(12,8))
plt.plot("Age","Annual Income (k$)", data = age)
plt.show()
```

### group by gender

```python
gender = mail_data.groupby("Gender").mean().reset_index()
gender.head()
```

```python
plt.figure(figsize=(12,8))
plt.bar("Gender","Annual Income (k$)", data = gender)
plt.show()
```
